```python
import pandas as pd
import random
import matplotlib.pyplot as plt

def game(player_1_n : int, player_2_n : int, rounds : int):
    
    '''
    Return the probabilities of winning for each player. 
    
    player_1_n (int): choice of number between 1 and 6;
    player_2_n (int): choice of number between 1 and 6;
    rounds (int): number of rounds played.
    
    '''
    player_1_victories = 0
    
    for _ in range(rounds):
        
        while True:
                        
            if random.randint(1, 6) == player_1_n:
                
                player_1_victories += 1
                
                break
                               
            else:
                                
                if random.randint(1, 6) == player_2_n:
                                       
                    break
            
    p_1 = player_1_victories / rounds
    
    p_2 = 1 - p_1
            
    return p_1, p_2

index, p_1, p_2, differential = [], [], [], []

for i in range(1, 1000):
    
    probabilities = game(3, 5, i)
    
    index.append(i)
    p_1.append(probabilities[0])
    p_2.append(probabilities[1])
    differential.append(probabilities[0] - probabilities[1])
    
plt.figure(figsize = (14, 6))

plt.plot(index, p_1, label = "Probability Player 1 wins")
plt.plot(index, p_2, label = "Probability Player 2 wins")
plt.title("Estimated Probability of Winning vs Number of Simulations")
plt.xlabel("Number of simulations (Rounds)")
plt.ylabel("Estimated probability")
plt.legend()

plt.show()
```
